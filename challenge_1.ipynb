```python
X = ttt.drop('class', axis=1) # all columns except class

y = ttt['class'] # class column
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
from tensorflow import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# create the model
input_shape=(9,)


model = Sequential()

model.add(Dense(32,
                activation='relu',
                input_shape=(9,)))

model.add(Dense(16,
                activation='relu'))

model.add(Dense(16,
                activation='relu'))

model.add(Dense(16,
                activation='relu'))

model.add(Dense(2,
                activation='softmax'))


```

```python
#compile
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy'])
```

```python
# train
history = model.fit(
    X_train,
    y_train,
    epochs=30,
    batch_size=32,
    validation_split=0.2
)
```

```python
#evaluate the model
test_loss, test_accuracy = model.evaluate(
    X_test,
    y_test
)
train_loss, train_accuracy = model.evaluate(
    X_train,
    y_train
)

print(f"train accuracy {train_accuracy} and loss {train_loss}")
print(f"test accuracy {test_accuracy} and  loss {test_loss}")
```
